# Q: what is the library used in implementing and dealing with the graph?
There are two libraries:

    NetworkX:
        NetworkX is a library primarily focused on graph theory and network analysis. It provides a wide range of tools for creating, manipulating, and analyzing graphs and networks.
        It supports various graph types, such as directed and undirected graphs, weighted graphs, and multigraphs.
        NetworkX is great for tasks like finding shortest paths, computing centrality measures, detecting communities, and visualizing graphs.
        It is well-suited for academic and research purposes in fields like social network analysis, biology, transportation, and computer science.
        NetworkX provides high-level graph operations, making it easy to work with complex network structures.
        It may not be as well-suited for general-purpose plotting beyond graph visualization, as it lacks some of the more advanced plotting capabilities found in Matplotlib.

    Matplotlib:
        Matplotlib is a comprehensive 2D plotting library widely used for data visualization. It provides a wide range of plotting functions for creating static, interactive, and animated visualizations.
        While it can plot graphs, it is not a graph theory-specific library like NetworkX. Instead, Matplotlib is more focused on general data visualization, including line plots, bar charts, scatter plots, histograms, and more.
        Matplotlib is highly customizable, allowing fine-tuning of visual elements like colors, labels, annotations, and axes appearance.
        It is widely used in various domains, such as data science, machine learning, finance, engineering, and research, where visualizing data in meaningful ways is essential.
        Matplotlib provides a broader range of plotting options beyond graph visualization, making it versatile for various visualization tasks.

In summary, NetworkX is specialized for graph theory and network analysis, while Matplotlib is a general-purpose data visualization library. NetworkX is suitable for working with graphs and networks, while Matplotlib is more versatile for creating different types of plots beyond graph visualizations. Depending on your specific needs, you may choose one library over the other or use them together to visualize graphs created with NetworkX using Matplotlib's extensive plotting capabilities.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

# Add edges
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 1)

# Plot the graph
pos = nx.spring_layout(G)  # Layout algorithm to position nodes
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=12, font_color='black', font_weight='bold', edge_color='gray', width=2)

plt.show()
```
